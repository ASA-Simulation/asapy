```python
import asapy
import time
```

#### 1. Login e autenticação

```python
asa = asapy.Asa(server="http://service.asa.dcta.mil.br")
```

#### 2. Listar simulações (obter Simulation Json e Components)

```python
sims = asa.simulations()
sims
```

Poderia salvar uma simulação manualmente também!

```python
my_sim = asapy.Simulation(name="my sim", duration=1000,
                          attributes={}, custom_data={}, station=asapy.Component(
                              identifier="AsaStation@AsaExtensions",
                              attributes={
                                  "bgRate": 10,
                                  "tcRate": 50,
                                  "enableDataLog": "",
                                  "enableDBRecorder": "//",
                                  "DBRecorderAsaMsgs": [
                                      "ASA_CUSTOM_DATA",
                                      "ASA_PLAYER_DAMAGED",
                                      "ASA_PLAYER_COLLISION",
                                      "ASA_PLAYER_CRASH",
                                      "ASA_PLAYER_KILLED",
                                      "ASA_WEAPON_RELEASED",
                                      "ASA_WEAPON_DETONATION"]},
                              subcomponents={"DBRecorderCustomMsgs": [
                                  {"identifier": "AsaCustomMsg1@AsaModels", "attributes": {}, "subcomponents": {}, "alias": {}}]},
                              alias={"alias1": "valor1",
                                     "alias2": "valor2"},
                              custom_data={"ws": {"key": "value"}}))

my_sim = asa.save_simulation(my_sim)
my_sim
```

Escolha de uma simulação

```python
sim = asa.simulation(1)
sim.to_df()
```

#### 3. Listar aliases

```python
doe = asapy.Doe()
aliases = doe.process_aliases_by_sim(sim, asa.component_configs())
aliases
```

#### 4. Edita parâmetros dos aliases

```python
aliases.loc['aliased_numTcThreads']['min'] = 3
aliases
```

#### 5. Criar o DOE

```python
df = doe.create(aliases, samples=10)
df
```

#### 6. Listar Métricas

```python
metrics_df = doe.process_metrics(sim)
metrics_df.loc['team_metrics_da_alice']['metrics']
```

```python
metric = 'msl_remaining'
```

#### 7. Rodar um batch com auto stop e barra de progresso

Criar batch


```python
batch = asapy.Batch(label="batch_001", simulation_id=sim.id)
```

Enviar batch para o manager

```python
batch = asa.save_batch(batch)
```

Prepara o chunks para rodar o batch

```python
exps = doe.prepare_experiments(df)
executions = batch.run(exps=exps, metric=metric)
executions
```

```python
count = 0
while True:
    print(batch.status())
    time.sleep(2)
    if count == 10:
        break
    count += 1
```
